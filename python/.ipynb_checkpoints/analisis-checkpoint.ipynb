```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def correlacion_pearson(data_frame, columna1, columna2):
    return np.corrcoef(data_frame[columna1], data_frame[columna2])[0,1];
```

# Datasets
Para los experimentos se crean las siguientes familias de instancias para evaluar los distintos métodos.
1. **peor-caso-HS1**: Instancias de peor caso para la heuristica secuencial golosa por grado en G.
2. **mejor-caso-HS1**: Instancias de mejor caso para la heuristica secuencial golosa por grado en G.
3. **peor-caso-HS2**: Instancias de peor caso para la heuristica secuencial golosa por grado en H.
4. **mejor-caso-HS2**: Instancias de mejor caso para la heuristica secuencial golosa por grado en H.
5. **instancias-costo**: Instancias generadadas de forma aleatoria para medir la perfomance de los algoritmos.
6. **instancias-calidad**: Instancias provistas por la cátedra para medir la calidad de los algoritmos

# Métodos
Los métodos evaluados en este notebook son los siguientes:
- **HS1**: Heuristica golosa constructiva secuencial por grado en G
- **HS1B**: Heuristica golosa constructiva secuencial por grado en G con busqueda local
- **HS2**: Heuristica golosa constructiva secuencial por grado en H
- **HS2B**: Heuristica golosa constructiva secuencial por grado en H con busqueda local
- **HGA**: Heuristica golosa coloreando G por las aristas de H
- **HGV**: Heuristica golosa coloreando G por los vertices de H
- **TSS**: Tabú search utilizando memoria por soluciones y vecindad con operador change y swap.
- **TSE**: Tabú search utilizando memoria por estructura de solución y vecindad con operador change y swap.
- **TSSA**: Tabú search utilizando memoria por soluciones y vecindad por coloreo en H. 
- **TSSB**: Tabú search utilizando memoria por soluciones, vecindad con operador change y swap y partiendo de una solucion inicial menos óptima.
- **TSEB**:Tabú search utilizando memoria por estructura de solución, vecindad con operador change y swap y partiendo de una solucion inicial menos óptima.
- **TSSAB**: Tabú search utilizando memoria por soluciones, vecindad por coloreo en H y partiendo de una solucion inicial menos óptima.

```python
# Comparamos los dataset de instancia-costo para cada heuristica contra el tiempo esperado en caso promedio.
df_complejidad_HS1 = df_resultados[(df_resultados["metodo"]=="HS1") & (df_resultados["dataset"]=="instancias-costo")].copy();
df_complejidad_HS1["tiempo_caso_promedio"] = ((df_complejidad_HS1["cantidad_vertices"])) * 0.00055 ;#Lineal
df_complejidad_HS1["tiempo_caso_peor"] = ((df_complejidad_HS1["cantidad_vertices"])**2.0) * 0.0000055 ;#Lineal
df_complejidad_HS2 = df_resultados[(df_resultados["metodo"]=="HS2") & (df_resultados["dataset"]=="instancias-costo")].copy();
fig = sns.scatterplot(data=df_complejidad_HS1, x='cantidad_vertices', y='tiempo');
fig = sns.scatterplot(data=df_complejidad_HS2, x='cantidad_vertices', y='tiempo');
fig = sns.lineplot(data=df_complejidad_HS1, x='cantidad_vertices', y='tiempo_caso_peor', color="red");
fig = sns.lineplot(data=df_complejidad_HS1, x='cantidad_vertices', y='tiempo_caso_promedio', color="orange");
fig.legend(labels=['O(n^2)','O(n)','HS1','HS2']);
fig.set(xlabel='cantidad_vertices', ylabel='tiempo (ms)');
plt.savefig("graficos/hs-complejidad.png")
```
